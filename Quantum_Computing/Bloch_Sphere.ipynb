```python
import numpy as np
from qiskit.visualization import plot_bloch_vector
import matplotlib.pyplot as plt

# Spherical coordinates: [r, theta, phi]
# r is usually 1
bloch_vectors = [
    [1, np.pi/2, 0],  # Qubit 0: H|0>
    [1, 0, 0],        # Qubit 1: X|1> -> |0>
    [1, 0, 0]         # Qubit 2: X|1> -> |0>
]

fig, axes = plt.subplots(1, 3, figsize=(12,4))
for i, vec in enumerate(bloch_vectors):
    plot_bloch_vector(vec, coord_type='spherical', ax=axes[i])
    axes[i].set_title(f'Qubit {i}')
plt.show()
```

```python
#2 Experiment with Spherical Coordinates for Bloch Sphere:
import numpy as np
import matplotlib.pyplot as plt

# Convert spherical to Cartesian
def spherical_to_cartesian(r, theta, phi):
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)
    return x, y, z

# Setup figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw Bloch sphere (wireframe)
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_wireframe(x, y, z, color='lightgray', linewidth=0.5, alpha=0.3)

# Define vectors (r=1 for Bloch sphere)
vectors = {
    "Positive X": (1, np.pi/2, 0),
    "Negative X": (1, np.pi/2, np.pi),
    "Positive Y": (1, np.pi/2, np.pi/2),
    "Negative Y": (1, np.pi/2, 3*np.pi/2),
    "Negative Z": (1, np.pi, 0),
}

# Plot vectors
for label, (r, theta, phi) in vectors.items():
    x, y, z = spherical_to_cartesian(r, theta, phi)
    ax.quiver(0, 0, 0, x, y, z, length=1.0, color='b', arrow_length_ratio=0.1)
    ax.text(x * 1.1, y * 1.1, z * 1.1, label, color='red')

# Formatting
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Bloch Sphere Vector Alignments")
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

# Simulator
sim = Aer.get_backend('statevector_simulator')

# ---- Case 1: |1> -> X ----
qc1 = QuantumCircuit(1)
qc1.x(0)  # apply X on |1>, so need to prepare |1> first
qc1.initialize([0,1], 0)  # set initial state to |1>
qc1.x(0)  # apply X gate

# Get statevector
result1 = execute(qc1, sim).result()
state1 = result1.get_statevector()

# ---- Case 2: |0> -> H ----
qc2 = QuantumCircuit(1)
qc2.h(0)  # apply H on |0>

# Get statevector
result2 = execute(qc2, sim).result()
state2 = result2.get_statevector()

# ---- Visualization ----
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plot_bloch_multivector(state1, title="|1> -> X -> |0>")

plt.subplot(1,2,2)
plot_bloch_multivector(state2, title="|0> -> H -> |+>")

plt.show()
```
